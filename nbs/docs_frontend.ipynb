# Front end (Streamlit app)

A Wealthsimple-inspired financial dashboard with an integrated chat interface. The dashboard auto-renders:

- **Top Summary** — Total balance, checking, savings, investments, net monthly cash flow
- **Net Worth Trend** — Smooth spline line chart over the last 12 months
- **Spending Breakdown** — Donut chart of spending by category for the current month
- **Income vs Expenses** — Grouped bar chart comparing the last 6 months
- **Budget Progress** — Color-coded progress bars (green/yellow/red) for each active budget
- **AI Insights** — Contextual alerts for overspending, unusual transactions, and savings suggestions

Below the dashboard, a chat input handles natural language queries for deeper exploration, what-if scenarios, budget creation, and data enrichment.

## How to run

1. Ensure your environment is configured (`.env` with `POSTGRES_*`, `PLAID_*`, and optionally `OPENAI_API_KEY` for agents).
2. From the repository root:
   ```bash
   pip install -e .
   streamlit run app/app.py
   ```
3. Open the URL shown in the terminal (typically http://localhost:8501).

## Chat interface

One chat input handles everything. Example prompts:
- **Spending by category:** "Show my spending breakdown", "Spending by category", "Pie chart of expenses"
- **Budgets:** "Budget overview", "Budgets over limit"
- **Spending over time:** "Spending over time", "Trend last 3 months"
- **Income vs expenses:** "Income vs expenses"
- **Recurring:** "Recurring payments", "My subscriptions"
- **Savings:** "Savings and investments"
- **Anomalies:** "Unusual transactions", "Anomaly detection"
- **What-if:** "What if I cut discretionary by 20%?"
- **Create budget:** "Create a monthly Coffee budget of 50 dollars"
- **Enrich:** "Enrich data" (run categorization and recurring/anomaly detection)

## Security

The Streamlit app uses the same database and environment variables as the notebooks. Run it only in a trusted environment (e.g. local or internal network). Do not expose it publicly without authentication.

## Run with Docker

You can run the Streamlit UI alongside the existing platform with Docker Compose:

1. Build the package and images (from the repo root):
   ```bash
   ./build-dist.sh   # or your existing build step that creates dist/*.tar.gz
   docker-compose build
   ```
2. Start the stack:
   ```bash
   docker-compose up
   ```
3. Access services:
   - JupyterLab: `http://localhost:8888`
   - Streamlit UI: `http://localhost:8501`

The `streamlit` service shares the same Postgres database (`db` service) and environment variables as the Jupyter `ipynb` service.
